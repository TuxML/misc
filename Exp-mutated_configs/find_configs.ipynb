```python
# Retrieve and print the size of the kernel from its cid in the old database.
def check_option_value(option, cid):
    try:
        socket = MySQLdb.connect(HOST, DB_USER, DB_PASSWD, "IrmaDB_prod_backup")
        cursor = socket.cursor()
        #query="SELECT cid FROM compilations WHERE compiled_kernel_version LIKE '4.15' and cid>29786 "
        query="SELECT cid, config_file, distrib_version, core_size, compressed_sizes, compilation_time \
                FROM `Compilations` where cid={}".format(cid)
        cursor.execute(query)
        result= cursor.fetchall()
        config=result[0][1]
        if (config is None):
            print ("Unable to retrieve cid=", str(cid))
            print(config)
        try:
            config_file = bz2.decompress(config).decode('ascii')
            #print(config_file)
        except Exception as e:
            print(str(e),"\n" + "Unable to decompress... ")
            exit(-1)

        for line in config_file.splitlines():
            if option in line:
                print(line, result[0][3],)

    except Exception as e:
        print(str(e),"\n" + "Unable to connect to database ")
        exit(-1)
    finally:
        cursor.close()
        socket.close()
```
